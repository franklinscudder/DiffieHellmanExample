```python
import Crypto.PublicKey.ECC as ecc
import matplotlib.pyplot as plt

point0 = ecc.generate(curve="p256").pointQ

pointsX = []
pointsY = []

for i in range(1, 100):
    point = point0 * i
    pointsX.append(int(point.x))
    pointsY.append(int(point.y))

plt.plot(pointsX, pointsY)
plt.show()
```
